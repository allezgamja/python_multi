# 데이터를 엑셀파일로 쓰기(.ExcelWriter)

1. pandas의 ExcelWriter 객체 생성
2. DataFrame 데이터를 지정된 엑셀시트(Sheet)에 쓰기
3. ExcelWriter 객체를 닫고, 지정된 엑셀파일 생성

```python
excel_exam_data1 = {'학생': ['A', 'B', 'C', 'D', 'E', 'F'],
                    '국어': [80, 90, 95, 70, 75, 85],
                    '영어': [90, 95, 70, 85, 90, 95],
                    '수학': [85, 95, 75, 80, 85, 100]}
df1 = pd.DataFrame(excel_exam_data1,columns=['학생','국어','영어','수학'] )
df1
```

```python
# ExcelWriter 객체 생성
excel_writer = pd.ExcelWriter('./data/학생시험성적2.xlsx', engine='xlsxwriter')
# Sheet에 데이터를 쓰기
df1.to_excel(excel_writer, index=False)
# ExcelWriter 객체를 닫고 지정된 엑셀파일 생성
excel_writer.save()
```

```python
excel_writer2 = pd.ExcelWriter('./data/학생시험성적3.xlsx', engine='xlsxwriter')
df1.to_excel(excel_writer2, index=False, sheet_name='중간고사')
excel_writer2.save()
```

```python
# 하나의 엑셀 파일에 두 개의 데이터 시트 생성
excel_exam_data2 = {'학생': ['A', 'B', 'C', 'D', 'E', 'F'],
                    '국어': [85, 95, 75, 80, 85, 100],
                    '영어': [80, 90, 95, 70, 75, 85],
                    '수학': [90, 95, 70, 85, 90, 95]}
df2 = pd.DataFrame(excel_exam_data2,columns=['학생','국어','영어','수학'] )
df2
```

```python
excel_writer3 = pd.ExcelWriter('./data/학생시험성적4.xlsx', engine='xlsxwriter')
df1.to_excel(excel_writer3, index=False, sheet_name='중간고사')
df2.to_excel(excel_writer3, index=False, sheet_name='기말고사')
excel_writer3.save()
```

# 엑셀파일 통합

효율적인 데이터 처리를 위해서는?
- 열의 header는 한 줄로, 데이터는 그 아래 입력
- 열 header나 데이터 입력부분에 셀병합기능은 이용하지 않는다.
- 하나의 셀에 숫자와 단위를 같이 쓰지 않는다.
- 하나의 열에 문자열, 숫자, 날짜 등 혼합하지 않는다.
- 데이터를 연도, 분기, 월, 제품별 등의 시트로 나누지 말고 가능하면 모든 데이터를 하나의 시트에 넣는다.

```python
# 파일지정
excel_data_files = ['./data/담당자별_판매량_Andy사원.xlsx',
                    './data/담당자별_판매량_Becky사원.xlsx',
                    './data/담당자별_판매량_Chris사원.xlsx']

# 각 파일에서 읽어온 데이터를 통합하기 위해 DataFrame 변수 하나 생성
total_data = pd.DataFrame()
```

```python
# 엑셀 파일을 하나씩 읽은 후 append()로 total_data변수에 차례대로 추가
for f in excel_data_files:
    df = pd.read_excel(f)
    total_data = total_data.append(df, ignore_index=True)

# ignore_index를 안하면 각자 원래의 인덱스대로 0,1,2,0,1,2...
    
total_data

# write: 앞에 내용에 덮어씌워진다
# append: 앞에 내용 건드리지 않고 추가됨
```

#### 지정된 패스의 모든 파일과 디렉토리 리스트 가져오기(glob)

# 여러 개의 엑셀 파일에서 데이터 수정(re)

```python
# 지정된 문자열의 패턴 이용하여 새로운 파일명으로 변경
import re

file_name = './data/담당자별_판매량_Andy사원.xlsx'

new_file_name = re.sub(".xlsx", "2.xlsx", file_name)
new_file_name
```

```python
# 여러 개의 엑셀파일을 각각 수정하고 새로운 파일명으로 저장
import glob
import re
import pandas as pd

# 원하는 문자열이 포함된 파일을 검색해 리스트 할당
excel_data_files1 = glob.glob("./data/담당자별_판매량_*사원.xlsx")

# 엑셀파일 수정
# 리스트에 있는 엑셀 파일만큼 반복수행
for f in excel_data_files1:
    # 엑셀파일에서 DataFrame형식으로 데이터 가져오기
    df = pd.read_excel(f)
    # 특정 열의 값 변경
    if(df.loc[1,'담당자']=='A'):    # 원소값지정
        df['담당자']='Andy'
    elif(df.loc[1, '담당자']=='B'):
        df['담당자']='Becky'
    elif(df.loc[1, '담당자']=='C'):
        df['담당자']='Chris'
        
    # 엑셀파일 이름에서 지정된 문자열 패턴을 찾아 파일명 변경
    f_new = re.sub(".xlsx", "2.xlsx", f)
    print(f_new)
    
    # 수정된 데이터를 새 파일명으로 저장
    new_excel_file = pd.ExcelWriter(f_new, engine='xlsxwriter')
    df.to_excel(new_excel_file, index=False)
    new_excel_file.save()
```
