## 3. 불(bool, 논리) 자료형
- True - 참, False - 거짓 (첫 글자만 대문자)
- 주로 제어문과 쓰인다.

```python
a = True
b = False
type(a)

# type(): 자료형 확인
```

```python
1 == 1
2 > 1
2 < 1
```

#### 자료형의 참과 거짓

```python
"python" # 참
""       # 거짓
[1,2,3]  # 참
[]       # 거짓
()       # 거짓
{}       # 거짓
1        # 참
0        # 거짓
None    # 거짓
```

```python
"" == True
# 빈 리스트 ""는 False라 True가 아니므로 False가 출력
```

```python
# while; 조건문이 참인 동안 조건문 안에 있는 문장을 반복해서 수행
# .pop(): 리스트의 마지막 요소를 끄집어내라
# 더이상 끄집어낼 것이 없어 리스트가 빈요소[]가 되면 거짓이 되어 멈춘다.
a = [1,2,3,4]
while a:
    print(a.pop())
```

```python
a = [1,2,3,4]
a
```

```python
if []:
    print("참")
else:
    print("거짓")
    
# 만약 []가 참이면 "참"을, []가 거짓이면 "거짓"을 출력하라
```

```python
if [1,2,3]:
    print("참")
else:
    print("거짓")

# [1,2,3]은 요소값이 있는 리스트이므로 참 -> "참"을 출력
```
