```python
import urllib.request as req    # http 웹서비스 담당해주는 모듈

local = "mushroom.csv"
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/mushroom/agaricus-lepiota.data'
req.urlretrieve(url, local)  # 데이터(url)를 특정 파일명(local)로 저장해라 
print("ok")
```

```python
# csv파일을 제대로 분석하려면 pandas!
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn import metrics #metrics 평가. 모듈객체
from sklearn.model_selection import train_test_split
```

```python
mr = pd.read_csv("mushroom.csv", header=None)
mr.head()
```

```python
label=[]
data=[]
attr_list=[]
```

```python
# for문 써서 행단위로 읽어오는 이유? 2차원으로 만들어주려고
for row_index, row in mr.iterrows():  #iterrows: 데이터내용을 행단위로 반복해서 계속 읽어라
    # print(type(row))   # pandas의 특징: 한 행을 가져오면 그걸 Series 데이터로 세로로 가져온다. 그래서 데이터 이상하게 가져와지면 type검사 항상 해라
    label.append(row.iloc[0]) # 행에 접근할때는 iloc 필요
    row_data=[]
    for v in row.iloc[1:]:    # 맨앞 0은 label이라서 냅둬야 한다. 그래서 1부터 시작
        row_data.append(ord(v))  # ord; 알파벳데이터를 아스키코드로 바꾸기
    data.append(row_data)  #여기까지는 1차원 
    
data
# a와 z는 다를 뿐인데, 수치로 바꾸면 서로 거리가 멀다고 인식할 수 있음. 그래서 a와b는 가깝고, a와z는 멀다고......
```

```python
data_train, data_test, label_train, label_test = train_test_split(data,label)
```

```python
clf = RandomForestClassifier()
clf.fit(data_train, label_train)
```

```python
predict = clf.predict(data_test)
ac_score = metrics.accuracy_score(label_test, predict)
cl_report = metrics.classification_report(label_test, predict)
print("정답률=", ac_score)
print("리포트=\n", cl_report)
```
