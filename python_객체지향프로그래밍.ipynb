- 구조적프로그래밍: 데이터가 하나라도 바뀌면 function을 수행하는 데 차질 생김
- 객체지향프로그래밍: 각각의 역할이 독립적으로 있는 역할중심
- 파이썬은 구조적, 객체지향 둘다 지원
- 함수가 데이터를 끌어안고 있으면 그에 대한 연산만 반복하므로 '인수'를 준다. 우리는 데이터가 아닌 코드만 재활용해야하므로.

![%EA%B0%9D2.png](https://user-images.githubusercontent.com/51535130/72500420-beb4b900-3877-11ea-8ef0-7f300af80f4b.png)

- 주기억장치 안에 데이터와 명령어들을 로딩시켜서 CPU에 의해 실행시키는 구조
- CPU에서 직접 주고받는 게 아니라 제어만 할 뿐, 직접적으로 주고받는 건 주기억장치
- 입출력(I/O)은 데이터의 이동이다. 해당 실행중인 프로세스 내 데이터를 기준으로 함. 만약 데이터가 바깥으로(다른 장치로) 나가면 출력, 바깥에서 들어오면 입력
- stream(스트림): 출력빨대 입력빨대 개념 - 일방향

![%EA%B0%9D3.png](https://user-images.githubusercontent.com/51535130/72500478-ec016700-3877-11ea-88e9-caecb2b1817c.png)

- 객체에서 중요한 것?
 - 은닉화(캡슐화, private): 오브젝트가 가진 주요 데이터는 감춰라. 만약 다른 곳에서 직접 접근하게 되면 해당 오브젝트의 데이터와 엉겨버리므로. 객체들 간 의사소통은 데이터가 아닌 객체 안의 함수로 해야 한다.
 - 추상화: 대놓고 보이지 말고 어떻게 했는지 가려라.

- Cal(1)은 데이터타입을 의미, 파이썬에서는 데이터타입 생략가능하므로 obj = new Cal()로 쓸 수 있다.
- obj(2, 객체)는 포인터일 뿐 실제로 수행하지 않는다. &100은 주소로, 직접 값을 넣는 게 아니라 값이 있는 주소를 입력한다.
- new(3)는 올리는 행위. Heap이라는 메모리공간 안에 내가 정해놓은 역할에 맞게 구성요소를 잡아준다.
- Cal(4)는 속성. 초기화는 속성에 한해서만 한다.

- static: 전역변수 저장
- stack: 지역변수 저장, 빨리 쓰고 버리는 데이터가 들어가는 곳
- x: 속성, 0: 속성값

- 배열 = Data * n
- 함수 = Code * n
- 객체 = (Data+코드) * n
- 배열, 함수, 객체 모두 다 주소(&100)를 다룬다.
