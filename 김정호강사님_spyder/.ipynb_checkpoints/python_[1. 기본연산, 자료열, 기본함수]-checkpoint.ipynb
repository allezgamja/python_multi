### 문자열 포맷 코드



```python
# 1. 숫자 바로 대입

>>> "I eat %d apples." % 3
'I eat 3 apples.
```

```python
# 2. 문자열 바로 대입

>>> "I eat %s apples." % "five"
'I eat five apples.'
```

```python
# 3. 숫자 값을 나타내는 변수로 대입

>>> number = 3
>>> "I eat %d apples." % number
'I eat 3 apples.'
```

```python
# 4. 2개 이상의 값 넣기

>>> number = 10
>>> day = "three"
>>> "I ate %d apples. so I was sick for %s days." % (number, day)
'I ate 10 apples. so I was sick for three days.'
```

```python
# 정렬과 공백
# 전체 길이가 10개인 문자열 공간에서 대입값을 오른쪽 정렬, 나머지는 공백으로
>>> "%10s" % "hi"
'        hi'

>>> "{0:>10}".format("hi")
'        hi'

# 왼쪽 정렬
>>> "%-10sjane." % 'hi'
'hi        jane.'

>>> "{0:<10}".format("hi")
'hi        '

# 가운데 정렬
>>> "{0:^10}".format("hi")
'    hi    '
```

```python
# 소수점 표현하기
>>> "%0.4f" % 3.42134234
'3.4213'


>>> y = 3.42134234
>>> "{0:0.4f}".format(y)
'3.4213'
```

```python
# 공백 채우기
>>> "{0:=^10}".format("hi")
'====hi===='
>>> "{0:!<10}".format("hi")
'hi!!!!!!!!'
```

```python
# f문자열 포매팅
>>> name = '홍길동'
>>> age = 30
>>> f'나의 이름은 {name}입니다. 나이는 {age}입니다.'
'나의 이름은 홍길동입니다. 나이는 30입니다.'
```

### 문자열 관련 함수

- 문자 개수 세기(count)
 - a = "hobby"
 - a.count('b')
 - 2
- 위치 알려주기1(find)
 - a = "Python is the best choice"
 - a.find('b') ----> b가 처음으로 나온 위치 반환
 - 14
 - a.find('k')
 - -1
- 위치 알려주기2(index) : find와 다른점은 문자열 안에 존재하지 않는 문자를 찾으면 오류가 발생
 - a = "Life is too short"
 - a.index('t')
 - 8
 - a.index('k')
- 문자열 삽입(join)
 - ",".join('abcd')
 - 'a,b,c,d'
- 소문자를 대문자로 바꾸기(upper)
 - a = "hi"
 - a.upper()
 - 'HI'
- 대문자를 소문자로 바꾸기(lower)
 - a = "HI"
 - a.lower()
 - 'hi'
- 왼쪽 공백 지우기(lstrip): 가장 왼쪽의 연속된 공백들을 모두 지운다.
 - a = " hi "
 - a.lstrip()
 - 'hi '
- 오른쪽 공백 지우기(rstrip)
 - a= " hi "
 - a.rstrip()
 - ' hi'
- 양쪽 공백 지우기(strip)
 - a = " hi "
 - a.strip()
 - 'hi'
- 문자열 바꾸기(replace)
 - a = "Life is too short"
 - a.replace("Life", "Your leg")
 - 'Your leg is too short'
- 문자열 나누기(split)
 - a = "Life is too short"
 - a.split() ---> 괄호 안에 값 없으면 공백을 기준으로 문자열 나눈다.
 - 나눈 값은 list 형태로 리턴
 - ['Life', 'is', 'too', 'short']
 - b = "a:b:c:d"
 - b.split(':')
 - ['a', 'b', 'c', 'd']

```python
>>> a = "Life is too short, You need Python"
>>> a[0]
'L'
>>> a[12]
's'
>>> a[-1]
'n'       # 뒤에서부터 셀 때는 0이 아닌 -1부터 센다.
```

```python
# 슬라이싱
>>> a[19:]      # 20번째부터 문자열 끝까지
'You need Python'

>>> a[:17]
'Life is too short' # 처음부터 18번째까지
```

# 기본함수
- 함수 설명: () 안에서 shift+tab
- print()
 - default가 end=\n(엔터)이므로 한 줄에 쓰려면 end=" "해줘야 한다.
 - print({},{}.format(a,b)): {} 안으로 format안의 a, b 값이 차례대로 대입된다.
 - print()나 print(i,end="\n"): 줄 단위가 바뀐다. print()는 default값이 enter.
 - print(i.end="\t"): 탭공간만큼 띄워서 가로로 출력된다.

- and / or
- in / not in
 - 1 in [1,2,3] >>> True
 - 1 not in [1,2,3] >>> false
- range(a,b): range는 b 바로 전의 숫자까지만 인식

- len(a): 문자열 a의 길이 구하기

- int(a): 값을 정수값으로 바꿔준다.
 - a = '20'
 - 1 + a = 오류 / 1 + int(a) = 21
